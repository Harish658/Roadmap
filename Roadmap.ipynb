```python
import osmnx as ox
```

```python
admin = ox.geocode_to_gdf('Budapest')
admin_poly = admin.geometry.values[0]
admin_poly
```

```python
G = ox.graph_from_polygon(admin_poly)
```

```python
G.number_of_nodes(), G.number_of_edges()
```

```python
nodes, edges = ox.graph_to_gdfs(G)
```

```python
import matplotlib.pyplot as p
p.style.use('dark_background')
```

```python
f, ax = p.subplots(1,1,figsize = (20,20))
edges.plot(ax=ax, edgecolor = 'red', linewidth = 0.2)
ax.axis('off')
```
